```python
#Features for Stress prediction:
stress_features = [
    'Age', 'Gender', 'Academic_Year', 'CGPA', 'Scholarship',
    'Upset', 'No_Control', 'Stressed', 'Overwhelmed', 'Confident',
    'Academic_Control', 'Irritation_Control', 'Top_Performance',
    'Angered_By_Grades', 'Piled_Up_Difficulties'
]
```

```python
#Features for Depression prediction:
depression_features = [
    'Age', 'Gender', 'Academic_Year', 'CGPA', 'Scholarship',
    'No_Interest', 'Hopeless', 'Sleep_Trouble', 'Low_Energy',
    'Appetite_Changes', 'Low_Self_Worth', 'Concentration_Problems',
    'Movement_Changes', 'Self_Harm_Thoughts'
]
```

```python
#Split the dataset for training and testing
from sklearn.model_selection import train_test_split

# for Stress
X_stress = df[stress_features]
y_stress = df['Stress_Level']

X_stress_train, X_stress_test, y_stress_train, y_stress_test = train_test_split(
    X_stress, y_stress, test_size=0.2, random_state=42, stratify=y_stress
)
print(f"Training set size for stress: {X_stress_train.shape}, Testing set size: {X_stress_test.shape}")

# for Depression
X_depression = df[depression_features]
y_depression = df['Depression_Level']

X_depression_train, X_depression_test, y_depression_train, y_depression_test = train_test_split(
    X_depression, y_depression, test_size=0.2, random_state=42
)
print(f"Training set size for depression: {X_depression_train.shape}, Testing set size: {X_depression_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
#Stress Model Training for Decision Tree Classifier
stress_model = DecisionTreeClassifier(random_state=42)
stress_model.fit(X_stress_train, y_stress_train)
y_stress_pred = stress_model.predict(X_stress_test)

print("Stress Model Accuracy:", accuracy_score(y_stress_test, y_stress_pred))
print("Stress Model Report:\n", classification_report(y_stress_test, y_stress_pred))
```

```python
#Stress Model Training for Random Forest Classifier
stress_model = RandomForestClassifier(n_estimators=300,criterion='entropy',
                             max_features='sqrt',min_samples_leaf=10,random_state=100).fit(X_stress_train, y_stress_train)
y_stress_pred = stress_model.predict(X_stress_test)

print("Stress Model Accuracy:", accuracy_score(y_stress_test, y_stress_pred))
print("Stress Model Report:\n", classification_report(y_stress_test, y_stress_pred))
```

```python
#Stress Model Training for Gradien tBoosting Classifier
stress_model = GradientBoostingClassifier(random_state=42)
stress_model.fit(X_stress_train, y_stress_train)
y_stress_pred = stress_model.predict(X_stress_test)

print("Stress Model Accuracy (Gradient Boosting):", accuracy_score(y_stress_test, y_stress_pred))
print("Stress Model Report (Gradient Boosting):\n", classification_report(y_stress_test, y_stress_pred))
```

```python
#Stress Model Training for Logistic Regression
stress_model = LogisticRegression(random_state=42, max_iter=1000)
stress_model.fit(X_stress_train, y_stress_train)
y_stress_pred = stress_model.predict(X_stress_test)

print("Stress Model Accuracy:", accuracy_score(y_stress_test, y_stress_pred))
print("Stress Model Report:\n", classification_report(y_stress_test, y_stress_pred))
```

```python
#Stress Model Training for GaussianNB
stress_model = GaussianNB()
stress_model.fit(X_stress_train, y_stress_train)
y_stress_pred = stress_model.predict(X_stress_test)

print("Stress Model Accuracy (Naive Bayes):", accuracy_score(y_stress_test, y_stress_pred))
print("Stress Model Report (Naive Bayes):\n", classification_report(y_stress_test, y_stress_pred))
```

```python
#Stress Model Training for KNN
stress_model = KNeighborsClassifier(n_neighbors=5)
stress_model.fit(X_stress_train, y_stress_train)
y_stress_pred = stress_model.predict(X_stress_test)

print("Stress Model Accuracy:", accuracy_score(y_stress_test, y_stress_pred))
print("Stress Model Report:\n", classification_report(y_stress_test, y_stress_pred))
```

```python
#Stress Model Training for SVM
stress_model = SVC(kernel='rbf', random_state=42)
stress_model.fit(X_stress_train, y_stress_train)
y_stress_pred = stress_model.predict(X_stress_test)

print("Stress Model Accuracy:", accuracy_score(y_stress_test, y_stress_pred))
print("Stress Model Report:\n", classification_report(y_stress_test, y_stress_pred))
```

```python
# Dictionary to store results
results = {}

#Decision Tree
model = DecisionTreeClassifier(random_state=42)
model.fit(X_stress_train, y_stress_train)
y_pred = model.predict(X_stress_test)
results["Decision Tree"] = [
    accuracy_score(y_stress_test, y_pred),
    precision_score(y_stress_test, y_pred, average='weighted'),
    recall_score(y_stress_test, y_pred, average='weighted'),
    f1_score(y_stress_test, y_pred, average='weighted')
]

#Random Forest
model = RandomForestClassifier(n_estimators=300, criterion='entropy',
                               max_features='sqrt', min_samples_leaf=10, random_state=100)
model.fit(X_stress_train, y_stress_train)
y_pred = model.predict(X_stress_test)
results["Random Forest"] = [
    accuracy_score(y_stress_test, y_pred),
    precision_score(y_stress_test, y_pred, average='weighted'),
    recall_score(y_stress_test, y_pred, average='weighted'),
    f1_score(y_stress_test, y_pred, average='weighted')
]

#Gradient Boosting Classifier
model = GradientBoostingClassifier(random_state=42)
model.fit(X_stress_train, y_stress_train)
y_pred = model.predict(X_stress_test)
results["Gradient Boosting"] = [
    accuracy_score(y_stress_test, y_pred),
    precision_score(y_stress_test, y_pred, average='weighted'),
    recall_score(y_stress_test, y_pred, average='weighted'),
    f1_score(y_stress_test, y_pred, average='weighted')
]

#Logistic Regression
model = LogisticRegression(random_state=42, max_iter=1000)
model.fit(X_stress_train, y_stress_train)
y_pred = model.predict(X_stress_test)
results["Logistic Regression"] = [
    accuracy_score(y_stress_test, y_pred),
    precision_score(y_stress_test, y_pred, average='weighted'),
    recall_score(y_stress_test, y_pred, average='weighted'),
    f1_score(y_stress_test, y_pred, average='weighted')
]

#GaussianNB
model = GaussianNB()
model.fit(X_stress_train, y_stress_train)
y_pred = model.predict(X_stress_test)
results["Naive Bayes"] = [
    accuracy_score(y_stress_test, y_pred),
    precision_score(y_stress_test, y_pred, average='weighted'),
    recall_score(y_stress_test, y_pred, average='weighted'),
    f1_score(y_stress_test, y_pred, average='weighted')
]

#KNN
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_stress_train, y_stress_train)
y_pred = model.predict(X_stress_test)
results["KNN"] = [
    accuracy_score(y_stress_test, y_pred),
    precision_score(y_stress_test, y_pred, average='weighted'),
    recall_score(y_stress_test, y_pred, average='weighted'),
    f1_score(y_stress_test, y_pred, average='weighted')
]

#SVM
model = SVC(kernel='poly', random_state=42)
model.fit(X_stress_train, y_stress_train)
y_pred = model.predict(X_stress_test)
results["SVM"] = [
    accuracy_score(y_stress_test, y_pred),
    precision_score(y_stress_test, y_pred, average='weighted'),
    recall_score(y_stress_test, y_pred, average='weighted'),
    f1_score(y_stress_test, y_pred, average='weighted')
]
df_results = pd.DataFrame(results, index=["Accuracy", "Precision", "Recall", "F1 Score"]).T
print("For Stress:")
print(df_results)
```

```python
warm_palette = sns.color_palette("OrRd", n_colors=df_results.shape[0])[::-1]
plt.figure(figsize=(15, 5))
sns.barplot(x=df_results.index, y=df_results["Accuracy"], hue=df_results.index, palette=warm_palette,  dodge=False, legend=False)
plt.ylabel("Accuracy")
plt.xlabel("Model")
plt.title("Model Accuracy Comparison for Stress Level")
plt.ylim(0, 1)
plt.show()
```
